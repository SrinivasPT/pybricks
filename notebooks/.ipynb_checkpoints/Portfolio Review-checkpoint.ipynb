```python
import matplotlib.pyplot as plt
from pybricks.pybricks.market_data.market_data import get_stock_history
```

```python
%pylab inline
```

```python
def plot_stocks(stocks):
    if (len(stocks) == 2):
        pylab.rcParams['figure.figsize'] = (16, 5)
        fig, axes = plt.subplots(nrows=1, ncols=2)
        axes_dict = {"0": axes[0], "1": axes[1]}
    elif (len(stocks) == 4):
        pylab.rcParams['figure.figsize'] = (16, 9)
        fig, axes = plt.subplots(nrows=2, ncols=2)
        axes_dict = {"0": axes[0, 0], "1": axes[0, 1], "2": axes[1, 0], "3": axes[1, 1]}
    
    for idx, stock in enumerate(stocks):
        plot_axes = axes_dict[str(idx)]
        price = get_stock_history(stock)
        price[['price', 'EMA20', 'EMA50']].plot(ax=plot_axes)
        plot_axes.set_title(stock)
```

```python
plot_stocks(['HDFC', 'HDFCBANK', 'ICICIBANK', 'BAJFINANCE'])
```

```python
plot_stocks(['CIPLA', 'DEEPAKNTR', 'IPCALAB', 'ALKEM'])
```

```python
plot_stocks(['ASIANPAINT', 'BERGEPAINT'])
```

```python
plot_stocks(['HINDUNILVR', 'LT', 'RELIANCE', 'TCS'])
```
