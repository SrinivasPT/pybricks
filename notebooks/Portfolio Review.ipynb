```python
import matplotlib.pyplot as plt
from pybricks.pybricks.market_data.market_data import get_stock_history, get_holdings, get_pnl, get_stocks_in_sector
```

```python
%pylab inline
```

```python
def plot_stock(stock):
    pylab.rcParams['figure.figsize'] = (16, 9)
    price = get_stock_history(stock)
    price[['price', 'EMA20', 'EMA50', 'EMA212']].plot()
```

```python
def plot_stocks(stocks):
    if (len(stocks) == 1):
        pylab.rcParams['figure.figsize'] = (16, 9)
        fig, axes = plt.subplots(nrows=1, ncols=1)
        axes_dict = {"0": axes[0]}
    elif (len(stocks) == 2):
        pylab.rcParams['figure.figsize'] = (16, 5)
        fig, axes = plt.subplots(nrows=1, ncols=2)
        axes_dict = {"0": axes[0], "1": axes[1]}
    else:
        number_of_stocks = len(stocks) + numpy.mod(len(stocks), 2)
        pylab.rcParams['figure.figsize'] = (16, number_of_stocks * 2 + 1)
        fig, axes = plt.subplots(nrows=number_of_stocks // 2, ncols=2)
        axes_dict = {"0": axes[0, 0], "1": axes[0, 1], "2": axes[1, 0], "3": axes[1, 1]}
        for idx in range(number_of_stocks):
            axes_dict[str(idx)] = axes[idx//2, numpy.mod(idx, 2)]
    
    for idx, stock in enumerate(stocks):
        plot_axes = axes_dict[str(idx)]
        price = get_stock_history(stock)
        price[['price', 'EMA20', 'EMA50', 'EMA212']].plot(ax=plot_axes)
        plot_axes.set_title(stock)
```

```python
def plot_stocks_for_sector(sector_name):
    stocks_in_sector = get_stocks_in_sector(sector_name)
    plot_stocks(stocks_in_sector)
```

```python
plot_stocks_for_sector('Pharma')
```
